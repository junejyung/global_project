```python
text = open('/content/drive/MyDrive/MSH[0,0]/HKDE/jobmustqual.txt').read()
text
```

```python
wordcloud = WordCloud().generate(text)
wordcloud  # 객체 주소값
wordcloud.words_
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.imshow(wordcloud)   # wordcloud 객체를 넣으면 워드클라우드 형태의 그래프 생성
plt.axis('off')  #눈금 삭제
plt.show()
```

```python
from wordcloud import STOPWORDS

STOPWORDS.add('Knowledge')
STOPWORDS.add('Skills')
STOPWORDS.add('experience')
STOPWORDS.add('experiences')
STOPWORDS.add('data')
STOPWORDS.add('Good')
STOPWORDS.add('related')
STOPWORDS.add('using')
STOPWORDS.add('working')
STOPWORDS.add('Engineering')
STOPWORDS.add('year')
```

```python
wordcloud = WordCloud(max_words=100,
                      stopwords=STOPWORDS).generate(text) # 'said' 사라짐
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
